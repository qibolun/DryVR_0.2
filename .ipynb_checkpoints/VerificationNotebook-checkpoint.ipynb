```python
args = {
	"vertex":["Acc1;Const","TurnLeft;Const","Acc1;Const","TurnRight;Const","Const;Const"],
	"edge":[[0,1],[1,2],[2,3],[3,4]],
	"variables":["car1_x","car1_y","car1_vx","car1_vy","car2_x","car2_y","car2_vx","car2_vy"],
	"guards":[
		"And(car1_y-car2_y>=10,car1_y-car2_y<=10.1)",
		"And(t>10.0,t<=12.0)",
		"And(car2_y-car1_y>=10,car2_y-car1_y<=10.1)",
		"And(t>10.0,t<=12.0)"
	],
	"resets":[
		"",
		"",
		"car1_vy=2.0",
		""
	],
	"initialSet":[[0.0,0.5,0.0,1.0,0.0,-17.0,0.0,1.0],[0.0,1.0,0.0,1.0,0.0,-16.5,0.0,1.0]],
	"unsafeSet":"@Allmode:And(car1_x-car2_x<2, car2_x-car1_x<2, car1_y-car2_y<2, car2_y-car1_y<2)",
	"timeHorizon":40.0,
	"bloatingMethod":"PW",
    "kvalue":[2,2,2,2,2,2,2,2]
}
```

```python
config = {
    "SIMUTESTNUM":2 # Number of hybrid simulations before verification to find counter-example
}
safety, reachtube = verify(args, TC_Simulate, paramConfig=config)
# verify the safey of the system
```

```python
from src.plotter.parser import parse

lines = reachtube.raw.split("\n")
initNode, y_min, y_max = parse(lines)
colors = ['blue', 'red', 'blue', 'yellow', 'black']

import matplotlib.pyplot as plt
import matplotlib.patches as patches
xdim = 0
ydim = [2,6]
idx = 0
fig1, ax1 = plt.subplots(2,5, sharey='row',sharex='col')
stack = [initNode]
modeList = ["Acc1","TurnLeft","Acc1","TurnRight","Const"]
while stack:
	node = stack.pop()
	for c in node.child:
		stack.append(node.child[c])
	lowerBound = []
	upperBound = []
	for key in sorted(node.lowerBound):
		lowerBound.append(node.lowerBound[key])
	for key in sorted(node.upperBound):
		upperBound.append(node.upperBound[key])

	for i in range(min(len(lowerBound),len(upperBound))):
		lb = map(float,lowerBound[i])
		ub = map(float,upperBound[i])
		#print lb[0],ub[0], lb[1],ub[1]

		for ci, d in enumerate(ydim):
			rect = patches.Rectangle((lb[xdim],lb[d]),ub[xdim]-lb[xdim],ub[d]-lb[d],color=colors[ci%len(colors)],alpha=0.7)
			ax1[0,idx].add_patch(rect)

	y_axis_min = min([y_min[i] for i in ydim])
	y_axis_max = max([y_max[i] for i in ydim])
	ax1[0,idx].set_title(modeList[idx],fontsize=12)
	ax1[0,idx].set_ylim([y_axis_min, y_axis_max])
	ax1[0,idx].plot()
	idx+=1


ydim = [1,5]
stack = [initNode]
idx = 0

while stack:
	node = stack.pop()
	for c in node.child:
		stack.append(node.child[c])
	lowerBound = []
	upperBound = []
	for key in sorted(node.lowerBound):
		lowerBound.append(node.lowerBound[key])
	for key in sorted(node.upperBound):
		upperBound.append(node.upperBound[key])

	for i in range(min(len(lowerBound),len(upperBound))):
		lb = map(float,lowerBound[i])
		ub = map(float,upperBound[i])
		#print lb[0],ub[0], lb[1],ub[1]

		for ci, d in enumerate(ydim):
			rect = patches.Rectangle((lb[xdim],lb[d]),ub[xdim]-lb[xdim],ub[d]-lb[d],color=colors[ci%len(colors)],alpha=0.7)
			ax1[1,idx].add_patch(rect)

	y_axis_min = min([y_min[i] for i in ydim])
	y_axis_max = max([y_max[i] for i in ydim])
	ax1[1,idx].set_ylim([-3.5, 0.5])
	ax1[1,idx].plot()
	idx+=1

plt.show()
```
